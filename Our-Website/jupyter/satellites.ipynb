### These columns have the following names and meanings:
<br>
<dl style="margin:0 30% 0 2%">
<dt>x1, y1, z1</dt>
<dd style="margin:0 20% 0 2%">The 3D Cartesian coordinates of the burst with origin on the surface of the earth. The x coordinate is the horizontal, the y coordinate is the depth and the z coordinate is the vertical component. Units are feet.</dd>
<dt>x2, y2, z2</dt>
<dd style="margin:0 20% 0 2%">The 3D Cartesian coordinates of the receiver with origin on the surface of the earth. The x coordinate is the horizontal, the y coordinate is the depth and the z coordinate is the vertical component. Units are feet.</dd>
<dt>alt1, alt2</dt>
<dd style="margin:0 20% 0 2%">The altitudes of the burst and receiver. The altitude is greater than or equal to the z coordinate because of the curvature of the earth. Units are feet.</dd>
<dt>horiz, vert</dt>
<dd style="margin:0 20% 0 2%">The horizontal and vertical offsets from the burst to the receiver. Units are feet.</dd>
<dt>slant</dt>
<dd style="margin:0 20% 0 2%">The slant range from the burst to the receiver. Units are feet.</dd>
<dt>angle</dt>
<dd style="margin:0 20% 0 2%">The angle of the burst to the receiver. Units are degrees. This angle is measured from the horizontal plane containing the burst location.</dd>
<dt>density1, density2</dt>
<dd style="margin:0 20% 0 2%">The air densities of the burst and receiver at alt1 and alt2. Units are grams per centimeter cubed.</dd>
<dt>rhor</dt>
<dd style="margin:0 20% 0 2%">The air depth (rhor) between the burst and receiver. Units are grams per centimeter squared.</dd>
</dl>

# Useful Functions

```python
linestyles = ['-', '--', '-.', ':', ':o']
   
def plotReset():
    plt.clf # clear figure
    plt.cla # clear axes
    plt.figure(figsize=(15,15)) # this also clears the previous plot
    
def plotScatter(xColumn, yColumn, df_ = res, color=None):
    if False: plt.scatter(env[xColumn], env[yColumn], alpha=1.0, zorder=2)
    if False: plt.plot(env[xColumn], env[yColumn], linestyle='dotted', alpha=0.8, zorder=1)
    plt.plot(df_[xColumn], df_[yColumn], ':o', alpha=1.0, color=color)

def plotFormats(title, xLabel, yLabel, legend, 
                xLim = (None,None), 
                yLim = (None,None),
                xTicks = (None,None,None), 
                yTicks = (None,None,None)):
    plt.xlabel(xLabel, fontsize=15, color='darkblue')
    plt.ylabel(yLabel, fontsize=15, color='darkblue')
    plt.xlim(xLim[0], xLim[1])
    plt.ylim(yLim[0], yLim[1])
    plt.title(title, fontsize=15, color='darkblue');
    plt.grid(True)
    plt.legend(legend[0], loc=legend[1])
    plt.xticks(fontsize=14, rotation=45)
    plt.yticks(fontsize=14, rotation=45)
    if xTicks[0] != None:
        plt.xticks(np.arange(xTicks[0], xTicks[1], step=xTicks[2]))
    if yTicks[0] != None:
        plt.yticks(np.arange(yTicks[0], yTicks[1], step=yTicks[2]))
    return
```

# Combine Altitudes and RHOR into One Graph

```python
query = res.iloc[:,:]

fig, ax1 = plt.subplots(figsize=(15,15))
ax1.set_title('\nAltitudes and RHOR\n', fontsize=15, color='Black')
ax1.set_xlabel('Time Stamp', fontsize=15)
ax1.set_ylabel('Altitude (feet)', fontsize=15, color='tab:red')
ax1.plot(query.time, query.alt1, linestyle='-', color='tab:red', alpha=1.0)
ax1.plot(query.time, query.alt2, linestyle='-', color='tab:green', alpha=1.0)
ax1.tick_params(axis='y', labelcolor='tab:red')
ax1.grid(True)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel('Air Depth (g/cm^2)\n', color='tab:blue', fontsize=15)  # we already handled the x-label with ax1
ax2.scatter(query.time, query.rhor, color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

ax1.legend(['Burst Altitude', 'Receiver Altitude'], loc="upper center")
ax2.legend(["RHOR Values"], loc="upper right")

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
